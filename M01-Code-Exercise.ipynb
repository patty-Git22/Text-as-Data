```python
import pandas as pd
```

```python
with open('2701-0.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

text = pd.DataFrame({
    'line_number': range(1, len(lines) + 1),
    'text': [line.strip('\n') for line in lines]
})

text.head(25)
```

### Question 1: How many lines does the file have?

```python
num_lines = len(text)
print(f'The total number of lines in the text is: {num_lines}')
```

### Question 2: At what line number does the novel’s contents begin?

```python
start_pattern = r'\*\** START OF THE PROJECT GUTENBERG'
start_line = text[text['text'].str.contains(start_pattern)].index[0] + 1
start_line_actual = start_line + 1
print(f'The actual starting line number of the novel is: {start_line_actual}')

#check
print(f"{start_line}: {text.iloc[start_line - 1]['text']}")
```

### Question 3: At what line number does the novel’s content end?
hint: The novel’s beginning and ending lines are announced in the Project Gutenberg boilerplate. Use string functions and/or regular expressions to find the answers

```python
end_pattern = r'\*\** END OF THE PROJECT GUTENBERG'
end_line = text[text['text'].str.contains(end_pattern)].index[0] + 1
print(f'The actual ending line number of the novel is: {end_line}')

#check
print(f"{end_line}: {text.iloc[end_line - 1]['text']}")
```
