```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import math
```

```python
letters = ['p', 't', 'k', 'a', 'i', 'u']
alpha = pd.Series(letters)
weights = [1/8, 1/4, 1/8, 1/4, 1/8, 1/8]
letter_weights = pd.DataFrame({'letters': letters, 'weights': weights})
letter_weights
```

```python
entropy = -sum([p * math.log(p, 2) for p in weights])
```

```python
L = 7
total_combinations = len(letters) ** L
print(f'Total combinations for length {L}: {total_combinations}')

X = [letters for i in range(L)]
library = pd.DataFrame(index = pd.MultiIndex.from_product(X)).reset_index()
library['word'] = library.apply(lambda x: x.str.cat(), axis=1)
library.word.head(30)
```

```python
def get_word(word_length = 7, weights = weights):
    word = alpha.sample(word_length, replace = True, weights = weights).str.cat()
    return word

def text_generate(n_words = 50, word_length = 7, weights = weights):
    words = [get_word(word_length, weights) for i in range(n_words)]
    text = ' '.join(words)
    return text

text_weighted = text_generate(n_words = 50, word_length = 7)
```

```python
large_text = text_generate(n_words=1000, word_length=7)

all_letters = large_text.replace(' ', '')
letter_counts = pd.Series(list(all_letters)).value_counts().sort_index()
letter_frequencies = letter_counts / len(all_letters)

print("Observed letter frequencies in generated text (1000 words):")
print(letter_frequencies)

print("\nExpected frequencies:")
for letter, freq in zip(letters, weights):
    print(f"{letter}: {freq:.4f}")
```

```python
comparison = pd.DataFrame({
    'letter': letters,
    'expected': weights,
    'observed': [letter_frequencies.get(letter, 0) for letter in letters]
})

comparison['difference'] = comparison['observed'] - comparison['expected']

print("\nComparison of Expected vs Observed Frequencies:")
print(comparison)
```

```python
def print_page(n_lines=40, words_per_line=10):
    for i in range(n_lines):
        line_text = text_generate(n_words=words_per_line, word_length=7)
        print(line_text)

print_page(n_lines=40, words_per_line=10)
```
